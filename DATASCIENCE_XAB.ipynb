# # Learn how to explore these data visually.
 

Matplotlib is a Python 2D plotting library that we can use to produce high quality data visualization. It is highly usable (as you will soon find out), you can create simple and complex graphs with just a few lines of codes!

Now let's load matplotlib to start plotting some graphs

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([0,1,2,3])
plt.show()
```

You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0, 1, 2, 3].

We can specify the x values separately. Let us try by mentioning the values.

```python
plt.plot([1,2,3,4],[1,4,6,9])
plt.show()
```

# Giving label to the x and y axis

We can label the xaxis and the yaxis which makes it easier to understand the data displayed in the graph.

```python
plt.plot([1,2,3,4],[1,4,6,9])

# Giving name to x axis and y axis 

plt.xlabel("XVALUES")
plt.ylabel("YVALUES")

#Give title for the graph
plt.title("MEAN TEMPERATURE OF JAIPUR")

plt.show()
```

we can also add title to the graph.

```python
plt.plot([1,2,3,4],[1,4,6,9])

# Giving name to x axis and y axis 

plt.xlabel("XVALUES",fontsize=12)
plt.ylabel("YVALUES",fontsize=12)

#Give title for the graph
plt.title("MEAN TEMPERATURE OF JAIPUR",fontsize=15)

plt.show()
```

# Change size of the graph

Using figsize

figsize() takes two parameters- width and height (in inches). By default the values for width and height are 6.4 and 4.8 respectively. 

Syntax:

plt.figure(figsize=(x,y)) 

```python
plt.figure(figsize=(10,10))
plt.plot([1,2,3,4],[1,4,6,9])

# Giving name to x axis and y axis 

plt.xlabel("XVALUES",fontsize=12)
plt.ylabel("YVALUES",fontsize=12)

#Give title for the graph
plt.title("MEAN TEMPERATURE OF JAIPUR",fontsize=15)

plt.show()
```

# Scatter plot


Scatter plots use a collection of points on a graph to display values from two variables. This allow us to see if there is any relationship or correlation between the two variables.

Let's see how mean temperature changes over the years!

```python
x=dtx.date
y=dtx.mean_temperature

plt.xlabel("Date")
plt.ylabel("MeanTemperature")
plt.title("MeanTemerature of Jaipur City")
plt.scatter(x,y)
plt.show()
```

Do you see that the x axis is filled with a thick line, and that there's no tick label available? This makes us unable to analyze the data.

Let's try to modify this scatter plot so that we can see the ticks!

Choose only several ticks
One reason why there's a thick bar below the x axis is that there are numerous labels (2 years daily data) on the x axis.

The first thing we are going to do is to then reduce the number of ticks/ points for the x axis. We do this using the np.arrange function as below:

```python
import numpy as np
x=dtx.date
y=dtx.mean_temperature

plt.xlabel("Date")
plt.ylabel("MeanTemperature")
plt.title("MeanTemerature of Jaipur City")
plt.scatter(x,y)

plt.xticks(np.arange(0,731,180))
plt.show()
```

See, now you can see the numbers clearer, but they are still overlapping.



# Task 8: Change x ticks interval so that you can see the dates clearly

```python
import numpy as np
x=dtx.date
y=dtx.mean_temperature

plt.xlabel("Date")
plt.ylabel("MeanTemperature")
plt.title("MeanTemerature of Jaipur City")
plt.scatter(x,y)
#arange function used to change the interval
plt.xticks(np.arange(0,600,100))
plt.show()
```

# Task 9: Rotate our x ticks label so that we can see more ticks more clearly


```python
import numpy as np
x=dtx.date
y=dtx.mean_temperature

plt.xlabel("Date")
plt.ylabel("MeanTemperature")
plt.title("MeanTemerature of Jaipur City")
plt.scatter(x,y)

#arange function used to change the interval
plt.xticks(np.arange(0,600,100))

#rotation function used to the values in desirable angle 
plt.xticks(rotation=30)
plt.show()
```

Good! Now, we can also change the size of the plot and the size of the ticks

```python
import numpy as np
x=dtx.date
y=dtx.mean_temperature

#figsize is used to increase the plot size
plt.figure(figsize=(10,10))

# Add x and y labels and set a font size

plt.xlabel("Date",fontsize=14)
plt.ylabel("MeanTemperature",fontsize=14)
# Add  title and set a font size

plt.title("MeanTemerature of Jaipur City",fontsize=20)

#Fontsize of x & y values is set
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.xticks(np.arange(0,600,100))
plt.xticks(rotation=30)
plt.scatter(x,y)
plt.show()
```

# Task 12: Change your marker shape!

```python
plt.figure(figsize=(10,10))
#to change the marker  shape
plt.scatter(x,y,marker="*")

plt.xticks(np.arange(0,731,60))
plt.xticks(rotation=30)
plt.xticks (fontsize = 12)
plt.yticks (fontsize = 12)
plt.xlabel("Date",fontsize=24)
plt.ylabel("MeanTemperature",fontsize=24)
plt.title("MeanTemperature in Jaipur",fontsize=30)
plt.show()
```

# Changing color of marker

```python
plt.figure(figsize=(10,10))

#to change the color of marker
plt.scatter(x,y,c='violet',marker='*')

plt.xticks(np.arange(0,731,80))
plt.xticks(rotation=30)
plt.xticks (fontsize = 19)
plt.yticks (fontsize = 19)
plt.xlabel("Date",fontsize=24)
plt.ylabel("MeanTemperature",fontsize=24)
plt.title("MeanTemperature in Jaipur",fontsize=30)
plt.show()
```
