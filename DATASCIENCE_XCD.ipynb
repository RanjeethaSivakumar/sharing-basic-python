# # Learn how to explore these data visually.
 

Matplotlib is a Python 2D plotting library that we can use to produce high quality data visualization. It is highly usable (as you will soon find out), you can create simple and complex graphs with just a few lines of codes!

Now let's load matplotlib to start plotting some graphs

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([0,1,2,3])
plt.show()
```

You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0, 1, 2, 3].

We can specify the x values separately. Let us try by mentioning the values.

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

# Giving label to the x and y axis

We can label the xaxis and the yaxis which makes it easier to understand the data displayed in the graph.

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("XVALUES")
plt.ylabel("YVALUES")
plt.show()
```

# we can also add title to the graph.

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("XVALUES")
plt.ylabel("YVALUES")
plt.title("Data Graph")
plt.show()
```

# Change size of the graph

Using figsize

figsize() takes two parameters- width and height (in inches). By default the values for width and height are 6.4 and 4.8 respectively. 

Syntax:

plt.figure(figsize=(x,y)) 

```python
plt.figure(figsize=(10,10))
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("XVALUES")
plt.ylabel("YVALUES")
plt.title("Data Graph")
plt.show()
```

# Scatter plot


```python
import matplotlib.pyplot as plt
import numpy as np
x=dt.date# x axis values is taken from date of jaipurcleandata csv file 
y=dt.mean_temperature # yaxis values is taken from mean_temperature  of jaipurcleandata csv file
plt.scatter(x,y)
plt.show()
```

Do you see that the x axis is filled with a thick line, and that there's no tick label available? This makes us unable to analyze the data.

Let's try to modify this scatter plot so that we can see the ticks!

Choose only several ticks
One reason why there's a thick bar below the x axis is that there are numerous labels (2 years daily data) on the x axis.

The first thing we are going to do is to then reduce the number of ticks/ points for the x axis. We do this using the np.arrange function as below:

```python
import matplotlib.pyplot as plt
import numpy as np
x=dt.date # x axis values is taken from date of jaipurcleandata csv file 
y=dt.mean_temperature # yaxis values is taken from mean_temperature  of jaipurcleandata csv file
plt.scatter(x,y)
# np.arange function is used to reduce the no of x axis /y axis values
# np.arange(start,stop.interval),more the interval lessee will be the values
plt.xticks(np.arange(0,700,180))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=dt.date
y=dt.mean_temperature
plt.scatter(x,y)
plt.xticks(np.arange(0,700,180))
#
plt.xticks(rotation=30)
plt.show()
```

```python

# Change the default figure size
plt.figure(figsize=(7,7))
x=dt.date
y=dt.mean_temperature
plt.scatter(x,y)
plt.xticks(np.arange(0,700,180))
plt.xticks(rotation=30)

# Add x and y labels, title and set a font size
plt.xlabel("Date",fontsize=14)
plt.ylabel("MeanTemperature",fontsize=14)
plt.title("MEAN  TEMPERATURE OF JAIPUR",fontsize=18)

plt.show()
```

# Displaying the x values in specified angle using rotation ()


```python

plt.figure(figsize=(7,7))
x=dt.date
y=dt.mean_temperature
plt.scatter(x,y)

plt.xticks(fontsize=12)# font size of xticks is specified
plt.yticks(fontsize=12)#font size of yticks is specified

plt.xticks(np.arange(0,700,180))

#Change x ticks interval so that you can see the dates clearly
plt.xticks(rotation=30)

plt.xlabel("Date",fontsize=14)
plt.ylabel("MeanTemperature",fontsize=14)
plt.title("MEAN  TEMPERATURE OF JAIPUR",fontsize=18)
plt.show()
```
